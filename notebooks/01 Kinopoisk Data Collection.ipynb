```python
def reviews_by_movie_id(movieId, limit = 250):
    url = f'https://api.kinopoisk.dev/v1.4/review?page=1&limit={limit}&movieId={movieId}'
    headers = {
        "accept": "application/json",
        "X-API-KEY": "ZMACHPY-D0AM2BS-J3Z30G6-H5B5S42"
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        reviews = pd.DataFrame(json.loads(response.text)['docs'])
        pages = json.loads(response.text)['pages']
        logger.info(f'{movie_id},{response.status_code}')
    else:
        logger.error(f'{movie_id},{response.status_code}')
        return reviews
    if pages > 1:
        for i in range(2, pages+1, 1):
            url = f'https://api.kinopoisk.dev/v1.4/review?page={i}&limit={limit}&movieId={movieId}'
            response = requests.get(url, headers=headers)
            reviews = pd.concat([reviews, pd.DataFrame(json.loads(response.text)['docs'])])
            logger.info(f'{movie_id}, {i}, of {pages}, {response.status_code}')
    return reviews
```
